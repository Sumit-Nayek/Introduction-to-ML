```python
## How to process multiple data into one file
import pandas as pd
import glob


# Get a list of CSV file names in a directory
file_list = glob.glob(
    'C:/Users/Sumit007/Desktop/Chandan Sir work/Introduction to ML and DL series/Data/Data for pre-processing/Years/*.csv')
## iterate through each of the files
for file in file_list:
    print(file)

# Create an empty DataFrame to store the data
all_data = pd.DataFrame()

# Loop through the list of CSV files and concatenate them
for file in file_list:
    df = pd.read_csv(file)
    all_data = pd.concat([all_data, df], ignore_index=True)

# Display the combined DataFrame
print(all_data)
type(all_data)
all_data.head(20)
```

```python
#### How to save csv from python
all_data.to_csv(
    'C:/Users/Sumit007/Desktop/Chandan Sir work/Introduction to ML and DL series/Data/Data for pre-processing/merged_data.csv'
    ,index=False)
```

```python
### Reading from the directory
## How to process multiple data into one file
import pandas as pd
import glob


# Get a list of CSV file names in a directory
file_list = glob.glob(
    'C:/Users/Sumit007/Desktop/Chandan Sir work/Introduction to ML and DL series/Data/Reading Excel files/Years/*.xlsx')
## iterate through each of the files
for file in file_list:
    print(file)

# Create an empty DataFrame to store the data
all_data = pd.DataFrame()

# Loop through the list of CSV files and concatenate them
for file in file_list:
    df = pd.read_excel(file)
    all_data = pd.concat([all_data, df], ignore_index=True)

# Display the combined DataFrame
print(all_data)
type(all_data)
all_data.head(20)
```

```python
#### Storing merged excel data into csv formate
all_data.to_csv(
    'C:/Users/Sumit007/Desktop/Chandan Sir work/Introduction to ML and DL series/Data/Reading Excel files/merged_data.csv'
    ,index=False)
```
