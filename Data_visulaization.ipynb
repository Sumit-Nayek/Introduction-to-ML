### Univariate cases

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for demonstration
np.random.seed(42) ## Fixing the seed to get same data when ever we run the code
data = np.random.normal(loc=0, scale=1, size=1000) ## Randomly generating 1000 values between 0 and 1
categories = ['A', 'B', 'C'] ## Creating category list
category_counts = [300, 400, 300] ## Frequancy of the categories

# Histogram
plt.hist(data, bins=30) ## Here we use the randomly generated data splited into 30 bin
plt.xlabel('Value') ## Setting the X label in the plot as 'Value'
plt.ylabel('Frequency') ## Setting the y label in the plot 'Frequancy'
plt.title('Histogram') ## Setting the title of the plot as 'Histogram'
plt.show() ## This code help us to see the output on the screen/ Just like print function
```

```python
# Bar Plot
plt.bar(categories, category_counts) ## Bar plot on categories and their frequancies
plt.xlabel('Category') ## Here x label is 'Category'
plt.ylabel('Count') ## y label is 'Count'
plt.title('Bar Plot') ## Title of the plot 'Bar Plot'
plt.show() ## It will show the output
```

```python
# Box Plot
plt.boxplot(data) ## Box plot of the randomly genrated data
plt.ylabel('Value') ## Here x label us 'Value'
plt.title('Box Plot') ## Title of the plot is 'Box Plot'
plt.show() ## Used for printing the output of a graphical object
```

```python
# Kernel Density Plot
import seaborn as sns
sns.distplot(data, hist=True, kde=True,
             bins=30, color = 'blue',
             hist_kws={'edgecolor':'black'}) ## ploting the density plot using seaborn librabry
plt.xlabel('Value') ## Set x label as 'Value'
plt.ylabel('Density') ## Set y label as 'Density'
plt.title('Kernel Density Plot') ## Seting title of the plot 'Kernel Density Plot'
plt.show()
```

```python
# Violin Plot
plt.violinplot(data, showmedians=True) ## ploting the voiling plot on the randomly generated data
plt.ylabel('Value') ## Here also the x label is 'Value'
plt.title('Violin Plot') ## Title of the plot is Violin Plot'
plt.show()
```

```python
## Pie digaram
colors = ['#ff9999', '#ffc000', '#8fd9b6'] ## List of colour
explode = (0, 0.1, 0.1)  # Explode the second slice

# Create the pie chart
plt.pie(category_counts, explode=explode, labels=categories, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) ## Pie diagram on categorical data
        ## Here explode define the space between the categories
        ## Here labels defines the list of category

# Add a title
plt.title('Fruit Distribution')

# Display the chart
plt.show()
```

### Bivariate cases

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.random.rand(100)
y = np.random.rand(100)
categories = ['A', 'B', 'C']
data = np.random.rand(100, len(categories))

# Scatter Plot
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

```python
# Box Plot
plt.boxplot(data, labels=categories)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

```python
# Violin Plot
plt.violinplot(data, showmedians=True)
plt.xticks(np.arange(1, len(categories) + 1), categories)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Violin Plot')
plt.show()
```

```python
# Bar Plot
mean_values = np.mean(data, axis=0)
plt.bar(categories, mean_values)
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.title('Bar Plot')
plt.show()
```
